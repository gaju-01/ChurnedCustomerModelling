```python
# splitting the data set into the training set and the test set
from sklearn.model_selection import train_test_split

# shuffling comes into picture when the data is ordered
X_train, X_test, y_train, y_test = train_test_split(
    data_set.iloc[:, 2:].values,
    data_set.iloc[:, 1:2].values,
    test_size=0.3,
    shuffle=True,
    random_state=34,
)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

# listing the indices of the column which are in need to  be transformed
indices = []
for i in range(len(X_train[0])):
    if type(X_train[0][i]) == str:
        my_LE = LabelEncoder()
        X_train[:, i] = my_LE.fit_transform(X_train[:, i])
        X_test[:, i] = my_LE.transform(X_test[:, i])
        indices.append(i)
print(indices)
my_col = ColumnTransformer(
    transformers=[("my_encoder", OneHotEncoder(), indices)],
    remainder="passthrough",
    transformer_weights={"my_encoder": 1},
)
X_train = my_col.fit_transform(X_train)
X_test = my_col.transform(X_test)
my_std_sclr = StandardScaler(with_std=True, with_mean=True)
X_train = my_std_sclr.fit_transform(X_train)
X_test = my_std_sclr.transform(X_test)

my_LE = LabelEncoder()
y_train = my_LE.fit_transform(y_train)
y_test = my_LE.transform(y_test)
```

```python
# now we just use the cross validation approach and SMOTE to produce the good results.....
# here we are doing the manual cross validation because if we just oevr smaple the data before the training
# set wour model  may produce the over opotimistic results . While doing the SMOTE as the proocess of
# cross validation ,then also we should be careful with selecting our classifier
from sklearn.model_selection import KFold
from imblearn.over_sampling import SMOTE
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier


def man_cross_val(classifier, co=1, approach=None, **params):
    my_smoter = SMOTE(random_state=34, k_neighbors=5)
    if approach is None:
        # shuffling comes into picture when the data is ordered
        approach = KFold(n_splits=5, shuffle=True, random_state=67)
    # getting the indices to split the data
    my_sens_table, my_pre_table = [], []
    for kf_tr_i, kf_val_i in approach.split(X_train, y_train):
        # splitting  the data set
        X_tra, y_tra = X_train[kf_tr_i], y_train[kf_tr_i]
        X_val, y_val = X_train[kf_val_i], y_train[kf_val_i]
        # Using the SMOTE technique to resmaple(Over sample) the data and visualising the data
        X_tra_os, y_tr_os = my_smoter.fit_resample(X_tra, y_tra)
        my_model = classifier
        # passing the parameters as arguments and using the classifier.set_params(**params) to set the
        # parameters of the model
        my_model = my_model.set_params(**params)
        my_model = my_model.fit(X_tra_os, y_tr_os)
        y_val_pred = my_model.predict(X_val)
        y_val_confu = confusion_matrix(y_val, y_val_pred)
        # calculating the sensitivity and precession rate of our model
        sensitivity, precession = y_val_confu[0][0] / (
            y_val_confu[0][0] + y_val_confu[0][1]
        ), y_val_confu[0][0] / (y_val_confu[0][0] + y_val_confu[1][0])
        my_sens_table.append(round(sensitivity * 100, 2))
        my_pre_table.append(round(precession * 100, 2))
    return (my_sens_table, my_pre_table)
```

```python
# printing the sample confusion matrix
li1 = []
Lr_model1 = LogisticRegression()
my_sens_table, my_pre_table = [], []

my_sens_table, my_pre_table = man_cross_val(
    Lr_model1, **{"random_state": 45, "solver": "newton-cg", "penalty": "l2"}
)
print(
    "Sensitivity:",
    np.array(my_sens_table).mean(),
    "\n",
    "Precession:",
    np.array(my_pre_table).mean(),
)
```

```python
Rf_model1=RandomForestClassifier()
my_sens_table1,my_pre_table1=[],[]
my_sens_table1,my_pre_table1=man_cross_val(Rf_model1,**{"n_estimators":100,"random_state":45,"max_depth":15})
print("Sensitivity:",np.array(my_sens_table1).mean(),"\n","Precession:",np.array(my_pre_table1).mean())
# Note:The sensitivity and precession rate that we have got is only for the specific score thershold
```

```python
# as you can see we have the results of both random forest classifier and Logistic Regression we can
# plot the PR curve for both of the logistic and random forest by the above reusult we can easily know that
# the optimal depth is 13 and optimal number of estimators is 85
Rf_model1.set_params(**{"random_state":34,"max_depth":13,"n_estimators":85})
y_test_pred=Rf_model1.predict(X_test)
my_cm=confusion_matrix(y_test,y_test_pred)
print(my_cm)
print("final sensitivity rate",(my_cm[0][0]/(my_cm[0][0]+my_cm[0][1]))*100)
print('final precession rate',(my_cm[0][0]/(my_cm[0][0]+my_cm[1][0]))*100)
# by the following we can see that random forest classifier produces good precession and recall rate
# which is satisfactory
```
